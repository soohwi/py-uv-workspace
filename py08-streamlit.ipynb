## 2. Streamlit - Inputs

1) app.py 수정

```python
%%writefile app.py

import streamlit as st

# 사용자 입력 관련 기능 소개
st.title('튜토리얼2: 입력 위젯')

# 문자열을 입력받을 수 있는 텍스트 필드 생성
name = st.text_input('이름을 입력하세요.')

# 숫자를 입력받는 위젯 - 최소/최대값 범위 지정
age = st.number_input('나이를 입력하세요', min_value=0, max_value=120)

# 드롭다운 형태로 항목 선택
hobby = st.selectbox('취미를 선택하세요', ['독서', '게임', '운동', '요리'] )

# 체크박스 - True/False 값 반환
agree = st.checkbox('개인정보 수집에 동의합니다')

# 입력값이 유효하고 동의했다면 결과 메시지 출력
if name and agree:
    st.success(f'{name}님, {age}세 / 취미: {hobby}')
```

```python
%%writefile app.py

import streamlit as st

# 버튼 및 조건부 로직 소개
st.title("튜토리얼 3: 버튼과 조건 처리")

# 버튼을 클릭하면 True를 반환
if st.button("누르세요"):
    # 애니메이션 효과를 표시
    st.balloons()
    # 버튼 클릭 메시지 출력
    st.write("버튼이 눌렸습니다!")

st.divider()

st.header("st.button", divider="rainbow")

if st.button("Reset", type="primary"):
    st.write("Reset is pushed")

if st.button("Say hello"):
    st.write("Hello there!! :smile:")
else:
    st.write("Goodbye")

st.divider()
st.button("Button",key=1) # 기본값
st.button("Button",key=2,use_container_width = True) # 부모의 너비 
st.button("Button",key=3,use_container_width = False) # 기본값
st.button("Button",key=4,disabled = True) # 비활성
st.button("Button",key=5,disabled = False) # 활성
st.button("Button",key=6,type = "secondary") # 기본값
st.button("Button",key=7,type = "primary") # 추가 강조

st.divider()
def handle_on_click():
    st.balloons()
    print("clicked on_click button")
def handle_on_click_args(*args):
    st.snow()
    print(f"clicked on_click args button with args={args}")
def handle_on_click_kwargs(**kwargs):
    st.toast("clicked!!!", icon="😍")
    print(f"clicked on_click kwargs button with kwargs={kwargs}")


st.button("on_click", on_click=handle_on_click)
st.button("on_click args", on_click=handle_on_click_args, args=("123",))
st.button("on_click kwargs", on_click=handle_on_click_kwargs, kwargs={"one":1})


```

## 3. Streamlit - File Upload

1. app.py 수정
* 실행 후 브라우저에서 <font color=red>Refresh</font> 클릭

* st.file_uploader(label, type=None, accept_multiple_files=False, key=None, help=None, on_change=None, args=None, kwargs=None, *, disabled=False, label_visibility="visible")

```python
%%writefile app.py

import streamlit as st
import pandas as pd

# 파일 업로드 기능 소개
st.title("📂 튜토리얼 4: 파일 업로드")

# 사용자가 파일을 업로드할 수 있는 위젯 (CSV만 허용)
uploaded_file = st.file_uploader("CSV 파일 업로드", type=["csv"])

if uploaded_file:
    # 업로드된 파일을 Pandas 데이터프레임으로 읽기
    df = pd.read_csv(uploaded_file)
    # 데이터프레임 상위 5개 행을 출력
    st.dataframe(df.head())


# accept_multiple_files=Tru: 여러개 이미지 업로드 가능
uploaded_files = st.file_uploader("파일을 선택해주세요", accept_multiple_files=True, type=["png","jpg","jpeg"])

for uploaded_file in uploaded_files:
    bytes_data = uploaded_file.read()
    # st.write(bytes_data)
    st.image(uploaded_file, caption=uploaded_file.name, use_container_width=True)


st.divider()
st.title('Download data')

st.header('Download as csv', divider='rainbow')
csv = pd.read_csv('./source_skn/jumsu.csv').to_csv()

st.download_button(
    label='Download data as CSV',
    data=csv,
    file_name='./source_skn/jumsu.csv',
    mime='text/csv'
)

with open('./source_skn/kingBob.jpeg', 'rb') as file:
    btn = st.download_button(
        label='이미지 다운로드',
        data=file,
        file_name='images/kingBob.jpg',
        mime='image/jpg'
    )
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns

# 차트 시각화 소개
st.title("📊 튜토리얼 5: 차트 시각화")

# 예시용 숫자 데이터 생성
data = pd.DataFrame({
    "x": [1,2,3,4,5],
    "y": [-0.5, 1, -1, 2 , 0]
})

# matplotlib로 선 그래프를 생성
fig, ax = plt.subplots()
ax.plot(data["x"], data["y"], marker='o')

# Streamlit 차트 렌더링
st.pyplot(fig)

# line chart
chart_data = pd.DataFrame(np.random.randn(10, 2), columns=["s", "t"])
st.line_chart(chart_data, width=0, height=300, use_container_width=True)

# bar chart
message = st.chat_message("assistant")
message.write("Hello human")
message.bar_chart(np.random.randn(30, 3))

# chart 사용 및 옵션
data = {"Year": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
"Sales": [100, 150, 200, 180, 150, 130, 200, 250, 270],
"Revenue": [50, 80, 120, 90, 100, 80, 150, 200, 220]}
df = pd.DataFrame(data)
sns.set_palette("Set2")           # 그래프의 기본 색상 테마를 'Set2'로 지정
fig = plt.figure(figsize=(10, 6)) # 다중라인 그래프 그리기

plt.title("Sales and Revenue Trend")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()                      # 범례
sns.lineplot(x="Year", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Year", y="Revenue", data=df, marker="o", label="Revenue")
plt.legend(loc='lower center')
st.pyplot(fig)


# pie chart
labels = ["A", "B", "C", "D"]
sizes = [random.randint(1, 100) for _ in range(len(labels))]

fig, ax = plt.subplots()
ax.pie(
    sizes, 
    labels=["A", "B", "C", "D"],
    colors= ["lightsteelblue", "thistle", "bisque", "lightsalmon"],  
    autopct="%1.1f%%"
    explode=[0 if s != min(sizes) else 0.1 for s in sizes] # 제일 작은 항목이 밖으로 튀어나오도록 하는 옵션
)

st.pyplot(fig)

```

```python
%%writefile app.py

import streamlit as st


# 사이드바와 조건부 표시 UI 소개
st.title("📑 튜토리얼 5: 사이드바와 토글")

# ------------------------------------------------------------------------------------
# 사이드바에 타이틀 출력
st.sidebar.title("사이드바 메뉴")

# 사이드바에 라디오 버튼으로 메뉴 선택 기능 추가
menu = st.sidebar.radio("메뉴 선택", ["홈", "정보", "설정"])

# 선택된 메뉴를 출력
st.sidebar.write("선택한 메뉴:", menu)

# 위의 소스에서 st.sidebar가 반복되는 부분을 with st.sidebar: 선언 후 사용하면 단축사용 가능
# with st.sidebar:
#     st.title("사이드바 메뉴")
#     menu = st.radio("메뉴 선택", ["홈", "정보", "설정"])
#     st.write("선택한 메뉴:", menu)
# ------------------------------------------------------------------------------------

# 토글 스위치를 통해 조건 제어
show = st.toggle("자세히 보기")

# 토글이 활성화되면 정보 메시지 출력
if show:
    st.info("여기에 자세한 내용이 표시됩니다.")
```

## 6. Streamlit - Session State

1) app.py 수정

* 실행 후 브라우저에서 <font color=red>Refresh</font> 클릭

```python
%%writefile app.py

import streamlit as st


# 세션을 활용하여 상태를 기억하는 방법 소개
st.title("튜토리얼 6: 세션 상태 관리")

# 세션 상태에 count가 없으면 초기값 설정
# 세션 상태에 저장을 안하고 그냥 count변수 설정으로 진행하면 버튼 클릭시 마다 초기화가 되어 값이 증가하지 않음
if "count" not in st.session_state:
    st.session_state.count = 0

# 버튼 클릭 시 count 값 증가
if st.button("카운트 증가"):
    st.session_state.count += 1

# 현재 상태 값 출력
st.write("현재 카운트:", st.session_state.count)
```
