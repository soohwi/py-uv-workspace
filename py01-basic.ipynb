# 자료형


1. 데이터 저장방법에 따른 자료형의 분류  
  직접(direct)형 : 직접 데이터를 표현하는 자료형 : int, float, complex      
  시퀀스(sequence)형: 다른 데이터를 포함하는 자료형, 순서가 있는 집합 자료형 : list,str, tuple, bytes, bytearray,range    
  매핑(Mapping)형 : 다른 데이터를 포함하는 자료형 : dict    
  집합(Set)형 : 순서가 없고, 중복된 값도 없다. : set, frozenset  


2. 변경 가능성에 따른 자료형 분류  
  변경 가능형(Mutable) : 데이터의 값을 변경할 수 있다.  : list, dict, set  
  변경 불가능형(Immutable) : 데이터의 값을 변경할 수 없다. : int, float, complex, str, tuple, frozenset


3. 데이터 저장 개수에 따른 자료형의 분류  
  리터널(Literal/Scalar)형 : 한 가지 객체만 저장한다. : int, bytes, str, float, complex  
  컨테이너(Container)형 : 여러 가지 객체를 저장한다. : list, tuple, dict, set, fronzenset

문자열(str)


문자열은 문자들을 표현하는 자료형으로 작은따옴표(')나 큰따옴표(")로 묶는다.  
문자열의 슬라이싱 : [인덱스 시작 : 인덱스 끝 : 스텝]  
스텝의 default은 1의 값을 갖는다.

```python
a = 'Life is too short, You need python'
print(a)

a = "Life is too short, You need python"
print(a)

a = "Python's favorite food is perl"
print(a)

# 문자열 그대로 사용하고 싶을 땐 역슬래시(\) 추가
a = 'Python\'s favorite food is perl'
print(a)

a = "\"Python\'s favorite food is perl\""
print(a)

# 여러 라인에 문자열을 넣을 땐 따옴표 3개 사용 (""", ''')
b = """
Life is too short,
You need python
"""
print(b)

# len() 문자열 길이
print(len(b))

# indexing(인덱싱)
s_val = 'hello python'
print(s_val[1])

# slicing(슬라이싱) [start:end:step] :step은 생략가능, default 1
print(s_val[1:3])       # 1부터 3미만
print(s_val[:])         # 전체
print(s_val[0:])        # 0부터 끝까지
print(s_val[-1])        # 맨 끝 값
print(s_val[-3: -1])    #
print(s_val[-1: -3])    # 생략된 step값이 1이기 때문에 -1 + 1 = 0 으로 n보다 뒷자리가 없어서 값 출력 안됨
print(s_val[-1: -3: -1])# 해당 방식으로 해야 원하는 -1부터 -3미만 으로 출력
```

```python
# (+) 연산자
# 다른 타입끼리 사용 불가
print(s_val + 2) # error
```

```python
# (*) 연산자
# 다른 타입끼리 사용 가능
print(s_val * 2) # hello pythonhello python # 문자열 s_val을 2번 반복
# 같은 타입끼리 사용 불가
print(s_val * s_val) #error
```

문자열 포매팅(서식지정) : 어떤 문서 양식을 만들어 넣고 빈칸에 필요한 내용을 채워서 문서를 지정한다.


문자열포매팅(서식지정) 형식


|코드|설명 |
|---|:---:|
|%d|정수|
|%f|실수|
|%s|문자열|
|%c|하나의문자|
|%o|8진수|
|%x|16진수|
|%%|%기호자체를출력하고자할때|

```python
s_val7 = 'python'
s_val8 = 83.5

# 일반적으로 이렇게 사용하진 않음
print('%s 의 점수는 %.1f입니다.' % ( s_val7, s_val8)) # python1, 2방식
print('{} 의 점수는 {}입니다.'.format(s_val7, s_val8)) # python3 방식
print('{1} 의 점수는 {0}입니다.'.format(s_val8, s_val7))
print('{subject} 의 점수는 {jumsu}입니다.'.format(jumsu=s_val8, subject=s_val7))
print('{{0}} 의 점수는 {{1}}입니다.'.format(s_val7, s_val8)) # { }자체를 출력할때 {{}}을 한다.
print('{{{0}}} 의 점수는 {{{1}}}입니다.'.format(s_val7, s_val8)) # {}자체와 값을 출력할때는 {{{}}}을 한다.


# <, > : 자릿수 확보 후 공백 위치 설정
print('{0:<10} 의 점수는 {1}입니다.'.format(s_val7, s_val8))  # 10자리 확보하고 남은 공간은 오른쪽을 여백으로 채운다.
#print('{:<10} 의 점수는 {}입니다.'.format(s_val7, s_val8))


print('{0:>10} 의 점수는 {1}입니다.'.format(s_val7, s_val8))
```

list는 (+) 연산자를 사용하여 연결하기를 할 수 있다.

```python
# list는 (+) 연산자를 사용하여 연결하기를 할 수 있다.
a = [1,2]
print(a + a)
# [1, 2, 1, 2]

# (*) 연산자를 사용하여 연결할 수 있음
print(a * 2)
# [1, 2, 1, 2]

# list는 mutable이므로 요소들을 변경할 수 있다.
b = ['apple', 'banana', 100, 200]
print(b)

# b[0] = 1      # b[0]값을 1로 변경
# b[1] = 2      # b[1]값을 2로 변경
b[0:2] = 1, 2   # 슬라이싱을 활용해 값을 한꺼번에 변경 가능
print(b)        # [1, 2, 100, 200]

b[0:2] = [50]   # 값을 배열안에 넣으면 0자릿수부터 2자릿수까지의 값을 50으로 변경하겠다는 의미
print(b)
# [50, 100, 200]

b[0:1] = [300, 400, 500]
print(b)
# [300, 400, 500, 'banana', 100, 200]

# 추가/삭제는 인덱싱이 아닌 슬라이싱으로 해야 함!
# 삭제
b[0:1] = []
print(b)
# ['banana', 100, 200]
```

list에서 확장 슬라이싱을 사용할 떄는 왼쪽과 오른쪽 요소 갯수가 같아야 한다.

```python
a = [2, 4, 6, 8]
print(a)

a[:] = [-1, -2, -3, -4]
print(a)
# [-1, -2, -3, -4]

a = [2, 4, 6, 8]
a[:] = [-1, -2, -3, -4, -5]
print(a)
# [-1, -2, -3, -4, -5]

# step값을 default인 1값이 아닌 다른 값을 지정하는 순간 요소 갯수가 같아야 함
a = [2, 4, 6, 8]
a[::2] = [-1, -2]
print(a)
# [-1, 4, -2, 8]

a = [2, 4, 6, 8]
a[::-1] = [-1, -2, -3, -4]
print(a)
# [-4, -3, -2, -1]

# 갯수가 일치하지 않을 경우 error
# ValueError: attempt to assign sequence of size 5 to extended slice of size 4
a = [2, 4, 6, 8]
a[::-1] = [-1, -2, -3, -4, -5]
print(a)
```

list는 한 변수에서 데이터를 각각 꺼내는 언팩킹(unpacking) 지원을 한다.

```python
fruits = ['apple', 'mango', 'grape', 'banana']
print(fruits)

zero, one, two, three = fruits
print(zero)
# apple

# 변수 2개만 저장하고 싶다면 불필요 값에 _ 넣기
zero, one, _, _ = fruits
print(zero, one)
# apple mango

# 변수 안에 값을 묶어서 가져와야 할 경우 변수명 앞에 * 표시
# * 는 한번만 사용 가능. zero, *group, *group2 = fruits 안됨.
zero, *group = fruits
print(zero, group)
# apple ['mango', 'grape', 'banana']

print(type(zero), type(group))
# <class 'str'> <class 'list'>

*group, zero = fruits
print(group, zero)
# ['apple', 'mango', 'grape'] banana

zero, *group, three = fruits
print(zero, group, three)
# apple ['mango', 'grape'] banana
```

Tuple(튜플)
- 변경이 불가능(immutable)한 순서(sequence)가 있는 자료형이다.
- 튜플은 ()소괄호로 표시한다.

```python
t = (1,2,3)
print(t, type(t))

t = 1, 2, 3     # 튜플 소괄호 생략 가능하지만 권장하지 않음
print(t, type(t))

# 소괄호 안 값을 1개만 지정할 경우 type이 tuple이 아닌 int로 나옴
# tuple로 사용 시 뒤에 콤마(,) 추가 필요
t = (2)     # int
t = (2,)    # tuple
t = 2,      # tuple
print(t, type(t))
```
