메인 메모리(RAM)
- CPU가 사용하는 저장 장치로 고속의 접근을 지원
- 크기 대비 생산 단가가 높기 때문에 하드 디스크에 비해 상대적으로 소량의 저장공간을 제공
* 대량의 데이터는 하드디스크에 저장됨

하드디스크와 RAM의 차이
- 하드디스크
    - 비휘발성 저장 매체로, 컴퓨터가 꺼져도 데이터를 영구적으로 저장
    - 주로 프로그램, 문서, 미디어 파일과 같은 대량의 데이터를 장기간 보관하는 데 사용됨
    - 테라바이트(TB) 단위의 데이터 저장 가능
- 메모리(RAM)
    - 휘발성 저장 매체로, 전원이 차단되면 저장된 데이터가 사라짐
    - 현재 실행중인 프로그램과 프로세스의 데이터를 일시적으로 저장
    - 빠른 데이터 액세스를 통해 CPU 처리 효율을 높임
    - 하드디스크 대비 상대적으로 작은 용량(보통 기가바이트GB)을 제공함

가비지 컬렉션(Garbage Collection)
- 객체의 참조 횟수를 추적하여 객체가 더 이상 필요하지 않게 될 때(참조 횟수가 0이 될 때) 메모리에서 데이터를 해제함
- 파이썬 프로그래밍 언어에서 자동으로 이루어지는 중요한 메모리 관리 프로세스
