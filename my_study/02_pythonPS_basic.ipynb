```python
import random

class Account:
    account_count = 0

    def __init__(self, name, balance):
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []

        self.name = name
        self.balance = balance
        self.bank = 'SC은행'

        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        num1 = str(num1).zfill(3)   # 1 -> '1' -> '001'
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)

        self.account_num = f'{num1}-{num2}-{num3}'
        Account.account_count += 1
    
    @classmethod
    def get_account_num(cls):
        return cls.account_count
    
    def deposit(self, amount):
        if amount >= 1:
            self.deposit_log.append(amount)
            self.balance += amount

            self.deposit_count += 1
            if self.deposit_count % 5 == 0:
                self.balance = self.balance * 1.01
            
    def withdraw(self, amount):
        if self.balance > amount:
            self.withdraw_log.append(amount)
            self.balance -= amount
    
    def deposit_history(self):
        print('---- 입금내역 ----')
        for amount in self.deposit_log:
            print(amount)

    def withdraw_history(self):
        print('---- 출금내역 ----')
        for amount in self.withdraw_log:
            print(amount)

    def display_info(self):
        print(f'은행이름: {self.bank}')
        print(f'예금주: {self.name}')
        print(f'계좌번호: {self.account_num}')
        print(f'잔고: {self.balance:,}')

userList = []

userK = Account('KIM', 10000000)
userL = Account('LEE', 10000)
userP = Account('PARK', 18302)

userList.append(userK)
userList.append(userL)
userList.append(userP)

userK.deposit(20000)
userK.deposit(1900)

userK.withdraw(10000)

userK.deposit_history()
userK.withdraw_history()
```
