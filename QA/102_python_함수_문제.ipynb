[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pickup_even(l):
    return [i for i in l if i % 2 == 0]

pickup_even([3, 4, 5, 6, 7, 8])
```

[문제4]  
콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하세요.

convert_int("1,234,567")

[출력결과]   
1234567

```python

def convert_int(str):
    # return int(''.join(s.split(',')))
    s = str.replace(',', '')
    return int(s)

convert_int("1,234,567")
```

[문제4]  
세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라. 단 if 문을 사용해서 수를 비교하세요.   
max(10, 2, 5)

[출력결과]  
10

```python
def print_max(x, y, z):
    max_val = x     # 0 으로 시작할 때 음수일 경우 측정불가하여 입력값중 하나를 기준으로 잡음.

    if y > max_val:
        max_val = y
    if z > max_val:
        max_val = z

    return max_val

print(print_max(10, 2, 5))
print(print_max(-10, -2, -5))
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2
2 와 5의 배수가 아닌 합계 :  16



```python
def result(*data):
    cnt = 0
    total = 0
    
    for i in data:
        if i % 2 != 0 and i % 5 != 0:
            cnt += 1
            total += i

    print(f"2와 5의 배수가 아닌 갯수: {cnt}")
    print(f"2와 5의 배수가 아닌 합계: {total}")

result(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)
```

문제6~13) 아래 함수를 호출했을때 정상적으로 실행이 되도록 호출하는 함수를 구현하세요.

```python
# 문제6)
def calc(a, b, /, c, *, d):
    return a + b + c + d

calc(1, 2, 3, d=4)
```

```python
# 문제7)
def introduce(name, /, age, *, country):
    return f"{name}, {age} years old from {country}"

introduce('홍길동', 10, country='seoul')
```

```python
# 문제8)
def f(x, *args, y, **kwargs):
    return x, args, y, kwargs

f(1, 2, 3, y=4, z=5)
# f(1, y=2)
```

```python
# 문제9)
def display(*args, **kwargs):
    print("ARGS:", args)
    print("KWARGS:", kwargs)

display(1, 2, x=1, y=2, z=3)
```

```python
# 문제10)
def config(*, debug=False, verbose=False):
    return f"Debug={debug}, Verbose={verbose}"

config()
config(debug=True, verbose=True)
```

```python
# 문제11)
def info(name, /, *, age, country):
    return f"{name}, {age}, {country}"

info('홍길동', age=14, country='seoul')
```

```python
# 문제12)
def combine(a, /, b, *, c):
    return a + b + c

combine(1, 2, c=1)
combine(1, b=2, c=1)
```

```python
#문제 13)
def test(a, b=2, *args, c=3, **kwargs):
    return a, b, args, c, kwargs

test(1, 3, 4, 5, c=6, d=7, e=8)
```
