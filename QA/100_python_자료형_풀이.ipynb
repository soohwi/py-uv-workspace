문제17)  
a 리스트에서 중복 숫자를 제거하세요.    
a = [1, 1, 1, 2, 2, 3, 3, 3, 4, 4, 5]  
※ 집합 자료형의 요솟값이 중복될 수 없다는 특징을 사용

```python
a = [1, 1, 1, 2, 2, 3, 3, 3, 4, 4, 5]
aSet = set(a)     # a 리스트를 집합자료형으로 변환
b = list(aSet)    # 집합자료형을 리스트 자료형으로 다시 변환
print(b)          # [1,2,3,4,5] 출력
```
