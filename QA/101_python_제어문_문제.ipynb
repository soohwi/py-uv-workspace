 [문제10]

   2. 띠구하는 공식

     0이면 "원숭이", 1이면 "닭"  2이면 "개",    3이면 "돼지",

[문제12]
A 학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.

[70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

list comprehension을 사용하여 A 학급의 평균 점수를 구해 보자.

```python
scores = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

total = sum(s for s in scores)
avg = total / len(scores)

print(avg)
```

[문제13]
======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 1
메뉴명: 짜짜
가격: 3000
신규 메뉴 짜짜(이)가 등록되었습니다.
{'짜짜': '3000'}

======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 2

==========메뉴판==========
짜짜 : 3000원
----------------------------------
1. 수정 | 2. 삭제 | 3. 나가기
> 3

======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 3
프로그램을 종료합니다.

```python
foods = {}

while True:
    print("======음식점 메뉴 관리 프로그램======")
    print("1. 신규 메뉴 등록")
    print("2. 메뉴판 전체 보기")
    print("3. 프로그램 종료")

    try:
        num = int(input("> "))
    except ValueError:
        print("숫자를 입력해주세요.")
        continue
    
    
    if num == 1:
        # 1. 메뉴명을 입력받아 해당메뉴가 사전에 키로 존재한다면
        #    "이미 등록된 메뉴입니다"를 출력하세요
        # 2. 사전에 존재하지 않는 메뉴(key)라면 가격을 입력받아
        #    key:value쌍으로 맵핑하여 사전에 저장하세요.
        f_name = input("메뉴명: ")
        
        if f_name not in foods:
            price = int(input("가격: "))
            foods[f_name] = price
            print(f"신규 메뉴 {f_name}(이)가 등록되었습니다.")
        else:
            print("이미 등록된 메뉴입니다.")
        
        print(foods)

    elif num == 2:
        # 1. 모든 키를 뽑아내서 리스트로 만드세요.
        # 2. 해당 리스트를 반복문으로 순회하여 메뉴와 가격을 출력하세요.
        #  ex) 출력 예시: "김치찌개 : 7000원"
        if not foods:
            print("등록된 메뉴가 없습니다.")
            continue

        menulist = foods.keys()

        print("==========메뉴판==========")
        
        for menu in menulist:
            print(f"{menu}: {foods[menu]}원")
        print("----------------------------------")
        print("1. 수정 | 2. 삭제 | 3. 나가기")

        try:
            select = int(input("> "))
        except ValueError:
            print("숫자를 입력해주세요.")
            continue

        if select == 1:
            name = input("가격을 변경할 메뉴의 이름을 입력하세요: ")

            # 해당 메뉴가 존재하는지 확인
            if name in foods:
                newPrice = int(input("변경할 가격 입력: "))
                # 해당 메뉴의 가격을 newPrice로 변경
                foods[name] = newPrice
                print(f"{name}의 가격이 {newPrice}원으로 변경되었습니다.")
            else:
                print(f"{name}(은)는 등록된 메뉴가 아닙니다.")
        elif select == 2:
            name = input("삭제할 메뉴의 이름을 입력하세요: ")

            # 해당 메뉴가 존재하는지 확인
            if name in foods:
                del(foods[name])
                print(f"{name}(이)가 삭제되었습니다.")
            else:
                print(f"{name}(은)는 등록된 메뉴가 아닙니다.")
        # elif select == 3:
        #     continue
        else:
            print("잘못된 번호입니다. 다시 입력해주세요.")
    
    elif num == 3:
        print("프로그램을 종료합니다.")
        break

    else:
        print("잘못된 번호입니다. 다시 입력해주세요.")
```
