## 예외(exception)
- 코드를 실행하는 중에 발생한 오류이다.
- 오류가 발생하는 이유는 프로그램이 잘못 동작하는 것을 막기 위함이다.

## 예외처리
- 예외처리는 에러가 발생해되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조
```
 try:
    실행할 코드
 except:
    오류가 발생했을 때 처리하는 코드
 else:
    오류가 없을 때 처리하는 코드
 finally:
    무조건 처리하는 코드      
```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할 때 에러가 발생하면 디버깅하는 스타일이다.
- EAFP
   - 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
   - 출처 : https://docs.python.org/3/glossary.html


- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy

````python
a = 10 / 0
````

````python
# try 블록이 정상적으로 실행되지 않으면 else블록은 실행되지 않는다
# try 영역에서 예외가 발생했을 때: try -> except -> finally
# try 영역이 정상적으로 수행됐을 때: try  -> else -> finally
try:
    a = int('3')
    b = int('a')
    c = int('4')
except:
    print("숫자가 아닌 문자가 입력되었습니다.")
else:
    print(f"{a} + {b} + {c} = {a + b + c}")
finally:
    print("program end")

print('-'*40)

try:
    a = int('3')
    b = int('5')
    c = int('4')
except:
    print("숫자가 아닌 문자가 입력되었습니다.")
else:
    print(f"{a} + {b} + {c} = {a + b + c}")
finally:
    print("program end")
````

## 다중 except
- 다중 except 선언이 가능하다.
- except뒤에 오는 as는 try 블록에서 발생한 이름을 받아온다.
```
  try:
     실행할 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드  
  except:
     예외가 발생했을 때 처리하는 코드  
```

* except 옆에 올 수 있는 것: "클래스"
1. 파이썬이 기본 제공하는 예외 클래스 (내장 예외)
except ZeroDivisionError as e:
except ValueError as e:
except IndexError as e:
except FileNotFoundError as e:
except TypeError as e:

2. 직접 만든 예외 클래스 (사용자 정의 예외)
class UserClass(Exception):
    def __str__(self):
        return "허용하지 않는 데이터입니다."

````python
y = [10, 20, 30]

try:
    index, x = map(int, input('인덱스와 나눌 숫자를 입력하세요.').split())
    print(y[index] // x)
except ZeroDivisionError as e:
    print("숫자를 0으로 나눌 수 없습니다.", e)

except IndexError as e:
    print("잘못된 인덱스입니다.", e)

except Exception as e:
    print(e)
else:
    print("정상처리 되었습니다.")
finally:
    print("program end")
````

````python
class Bird:
    # @abstractmethod
    def fly(self):
        raise NotImplementedError

b = Bird()
b.fly()
````

인스턴스화 자체를 막느냐, 못 막느냐가 다름.

NotImplementedError     
→ 객체는 일단 만들어짐. fly()를 실제로 호출하는 순간에만 에러가 남. 그 전까지는 "구현 안 됐다"는 사실이 겉으로 드러나지 않음

abstractedmethod        
→  객체를 만드는 시점에서 바로 막힘. fly()를 안 만들었다는 걸 훨씬 일찍(코드 실행 초반에) 알 수 있음


* 왜 NotImplementedError 방식도 실무에서 쓰는지?        
abstractmethod가 더 엄격하고 안전한데도, NotImplementedError를 쓰는 이유:       

1. ABC를 상속받기 싫을 때: 이미 다른 클래스를 상속받고 있어서 다중 상속 구조가 복잡해지는 걸 피하고 싶을 때
2. "이 메서드는 나중에 자식이 오버라이드하면 좋다"는 느슨한 가이드 정도로만 쓰고 싶을 때 (강제는 아니고 권장)
3. 간단한 스크립트나 소규모 프로젝트에서 abc 모듈까지 끌어오기 부담스러울 때


    
==>
실수를 미리 잡고 싶다면 @abstractmethod + ABC가 훨씬 안전하고, 실무에서도 진짜 "이 클래스는 절대 단독으로 못 쓰게 하고 싶다"는 의도가 명확하면 abstractmethod를 선호


````python
class Bird:
    def fly(self):
        raise NotImplementedError   # 상속받은 자식클래스에서 fly()을 오버라이딩 해줘야 함
    
class Eagle(Bird):
    def display(self):
        print("display")
    def fly(self):
        print("very fast")

eagle = Eagle()
eagle.fly()
````

````python
from abc import abstractmethod

# class Bird: (ABC 상속 안 함)
# @abstractmethod를 붙였는데도 아무 효과가 없음. 
# Eagle()도 잘 만들어지고, eagle.fly()도 에러 없이 그냥 pass(아무 것도 안 함)만 실행되고 끝남
class Bird:
    @abstractmethod
    def fly(self):
        pass

class Eagle(Bird):
    def display(self):
        print('display')

#   def fly(self):
#     print('very fast')

eagle = Eagle()
eagle.fly()
````

````python
from abc import ABC, abstractmethod

# class Bird(ABC): (ABC 상속함)
class Bird(ABC):
    @abstractmethod
    def fly(self):
        pass

class Eagle(Bird):
    def display(self):
        print('display')

#   def fly(self):
#     print('very fast')

# eagle = Eagle() 이 줄에서 바로 에러. 
# fly()를 구현 안 한 채로는 Eagle 객체 자체를 만들 수도 없음
eagle = Eagle() #ABC을 상속받은 Bird 추상클래스이다. Bird을 상속받은 Eagle도 추상클래스임
eagle.fly()
````

````python
class UserClass(Exception):
    def __str__(self):
        return "허용하지 않는 데이터입니다."
    
try:
    x = int(input("숫자입력: "))
    if x < 0:
        raise UserClass
    print(x)

except UserClass as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e:      # 다중 except를 사용할 때 부모클래스는 제일 아래에 위치함
    print(e)
except BaseException as e:
    print(e)
````
