# 식별 연산자 (Identity Operators)
- 식별 연산자는 두 객체가 같은 객체인지를 확인한다.
- is : 두 객체가 같은 객체인지 확인
- is not 두 객체가 다른 객체인지 확인
- 결과 : True, False

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)  # True (a와 b는 같은 객체)
print(a is c)  # False (a와 c는 같은 값을 가지지만 다른 객체)

# is not : 멤버 연산자 not in 과 헷갈리지 않기
print(a is not c)
````

#   멤버 연산자 (Membership Operators)
- 멤버 연산자는 시퀀스 내에 값이 존재하는지 확인한다.
- in : 특정값이 멤버의 값인지 확인
- not in : 특정값이 멤버의 값인지 아닌지 확인
- 결과 : True, False

````python
a = [1, 2, 3]
print(2 in a)  # True

# not in : 식별 연산자 is not 과 헷갈리지 않기
print(4 not in a)  # True
````

[테스트1]
age=35

````python
# age = int(input('나이를 입력하세요.'))
# point_card = input('포인트카드를 소지하고 계신가요? [예, 아니오]')
# count = int(input('지금까지 몇편의 영화를 보셨나요?'))
# 
# if age < 18:
#     print('티켓을 판매할 수 없습니다.')
# elif 18 <= age < 60:
#     print('티켓은 8000원입니다.')
# elif age >= 60 or point_card == '예':
#     print('티켓은 5000원입니다.')

age = int(input('나이:'))
point_card = bool(input('포인트 소지여부:'))
count = int(input('영화 관람 티켓 갯수:'))

if age < 18 :
  print('티켓을 판매할 수 없습니다')
elif age >=60 or (point_card and count==5):
  print('티켓은 5000원입니다')
else:
  print('티켓은 8000원입니다')
````

````python
# 실무용
# ===== 상수 정의 (실무에서는 하드코딩된 숫자를 이렇게 이름 붙여서 관리) =====
MINOR_AGE_LIMIT = 18          # 미성년자 기준
SENIOR_AGE_LIMIT = 60         # 경로 우대 기준
LOYAL_VIEW_COUNT = 5          # 포인트 할인 기준 관람 횟수

REGULAR_PRICE = 8000
DISCOUNT_PRICE = 5000


def get_ticket_price(age: int, point_card: bool, count: int) -> str:
    """
    나이, 포인트카드 소지 여부, 관람 횟수를 받아
    티켓 가격 안내 메시지를 반환한다.
    """
    # 1. 미성년자는 판매 불가 (가장 먼저 걸러내는 게 실무 관례 - 예외/제약 조건을 앞에 둠)
    if age < MINOR_AGE_LIMIT:
        return '티켓을 판매할 수 없습니다'

    # 2. 할인 대상 여부 판단 (조건이 복잡하니 변수로 분리해서 가독성 확보)
    is_senior = age >= SENIOR_AGE_LIMIT
    is_loyal_customer = point_card and count >= LOYAL_VIEW_COUNT
    is_discounted = is_senior or is_loyal_customer

    # 3. 가격 결정
    price = DISCOUNT_PRICE if is_discounted else REGULAR_PRICE

    return f'티켓은 {price}원입니다'


# ===== 테스트 =====
if __name__ == '__main__':
    test_cases = [
        {'age': 35, 'point_card': True, 'count': 5},   # 기대: 5000원
        {'age': 17, 'point_card': True, 'count': 5},   # 기대: 판매 불가
        {'age': 35, 'point_card': True, 'count': 3},   # 기대: 8000원
    ]

    for i, case in enumerate(test_cases, start=1):
        result = get_ticket_price(**case)
        print(f'[테스트{i}] {case} → {result}')
````

````python
# while문을 강제적으로 빠져나왔기 때문에 else영역을 실행하지 않는다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
  if a==5:
    break   # 반복문을 강제적으로 빠져나올때 사용한다. 즉 반복문을 중단시킨다
else:
  print('end')
````

````python
#  continue는 다음 루프(조건식)로 바로 넘어간다.
# 반복문을 중단하지는 않으므로 else문이 실행된다.
a = 10 #5
while a > 5:
  a -= 1
  if a % 2 == 0 :
    continue   # 이동
  print(a)
else:
  print('end')
````

## enumerate 함수
enumerate(반복가능객체, start)
 - enumerate()함수는 인덱스와 객체를 동시에 저장할 수 있는 특징을 가진 for문을 좀더 효율적으로 사용하기 위한 함수이다.
 - enumerate()함수는 반복가능객체(list, tuple, str)을 입력으로 받아 인덱스를 포함하는 enumerate 객체를 리턴한다.
 - enumerate([순서자료형], [시작값=0])의 형식을 가지며 반복가능객체은 반복자가 가능한 객체가 입력되고 시작값은 입력시 해당수 부터 증가하며, 생략했을 경우 기본값은 0 이다.


````python
for k in enumerate(['apple','python','microsoft']):
  print(k)

print(f"{'-' * 30}")

for k, v in enumerate(['apple','python','microsoft']):
  print(k, v)

print(f"{'-' * 30}")

# enumerate(iterable, start) => start는 생략이 가능하고 생략하면 default가 0임
for k, v in enumerate(['apple','python','microsoft'], 1):
  print(k, v)
````

````python
a = {'one':10, 'two':20}
for k in enumerate(a):
  print(k)

print(f"{'-' * 30}")

for i, k in enumerate(a):
  print(i, k)

for i, k in enumerate(a):
  print(i, a[k])

print(f"{'-' * 30}")

# index가 필요없이 키값만 받아올 경우
for _, k in enumerate(a):
  print(a[k])
````

### zip
- 매개인자로 여러 목록을 받아 각 목록의 요소를 튜플에 담아 목록에 저장한다.

````python
a = [1, 2, 3, 4, 5]
b = ['a','b','c','d','f']

for x in zip(a,b):
  print(x)

print('-'*30)

# 갯수 매칭이 다르다면 짧은 기준으로 맞춰서 반환됨
a = [1, 2, 3, 4, 5]
b = ['a','b','c','d']

for y in zip(a,b):
  print(y)

print('-'*30)

# 묶어준 것에 대한 index가 필요하다면 enumerate 사용
a = [1, 2, 3, 4, 5]
b = ['a','b','c','d']

for i, x in enumerate(zip(a,b)):
   print(i, x)

for i, (k, x) in enumerate(zip(a,b)):
   print(i, k, x)
````

### List Comprehension
- List 정의시 값을 정하지 않고 호출시 리스트 내의 값들이 처리되도록 구성
- A = [표현식 for 항목 in 반복가능객체(iterable)]
- A = [표현식 for 항목 in 반복가능객체(iterable) if 논리식]
- A = [표현식 for 항목1 in 반복가능객체(iterable)1 for 항목2 in 반복가능객체(iterable)2 if 논리식]

````python
s = []
for x in range(10):
    s.append(x**2)
print(s)

# list comprehension 방식
# [  결과값(뭘 담을지)   for 변수 in 반복가능객체 ]
s = [ x**2 for x in range(10) ]

print(s)
````

````python
s = []
for x in range(10):
    if x % 2 == 0:
        s.append(x**2)
print(s)

# list comprehension 방식 (if문이 있는 경우)
# [  결과값(뭘 담을지)   for 변수 in 반복가능객체    if 조건(선택사항) ]
s = [x**2 for x in range(10) if x % 2 == 0]
print(s)
````

````python
s = []
for x in range(5):
    if x % 2 == 0:
        s.append(x)
    else:
        s.append(-x)
print(s)

# list comprehension 방식 (if else문이 있는 경우)
# [x**2 for x in range(10) if x % 2 == 0]           # if만 있으면 맨 뒤 (필터링용)
# [x**2 if x % 2 == 0 else 0 for x in range(10)]    # if-else 있으면 맨 앞 (값 결정용)

s = [x if x % 2 == 0 else -x for x in range(5)]
print(s)
````

````python
s = []
for i in range(1, 3):       # 1,2
    for j in range(4, 8):   # 4,5,6,7
        #print('({},{})'.format(i,j))
        #s.append('({},{})'.format(i,j))
        s.append((i,j))
print(s)

# comprehension
s = [(i, j) for i in range(1, 3) for j in range(4, 8)]
print(s)
````

````python
# 총판매금액을 구하는 프로그램(list comprehension을 사용)  
cnt = [1, 1, 2, 1]  
prod = [500, 1000, 400, 600]

s = []
for i, j in zip(cnt, prod):
    s.append(i * j)
print(s)

total = sum(s) # sum 내장함수: 반복값을 합계 내주는 내장함수
print(total)

# comprehension 사용
s = [ i * j for i, j in zip(cnt, prod)]
total = sum(s)
print(total)
````

### Generator Expression
 - 파이썬에는 tuple(immutable)에는 comprehension이 존재하지 않고
    syntax 상으로 generator 가 생성한다.
 - 값을 하나씩 “생성”하는 반복자(iterator). 메모리를 덜 사용하며, **지연 평가(Lazy Evaluation)**가 특징이다.
 - generator를 호출하면 next 메소드를 통해 하나씩 호출하며 StopIteration 이 발생하면 종료된다.

````python
squares = (x **2 for x in range(5))
print(squares)   # 이건 제너레이터 객체 생성 (실행 아님)
print(dir(squares))

type(squares.__next__)

# generator 는 반복객체로 for문에서 사용 가능
for i in squares:
    print(i)
````

````python
squares = (x **2 for x in range(3))
print(squares)   # 이건 제너레이터 객체 생성 (실행 아님)
print(dir(squares))

print(next(squares)) # 0
print(next(squares)) # 1
print(next(squares)) # 4
print(next(squares)) # StopIteration : 더이상 가져올 값이 없어서 error
````

````python
# 개인정리
squares1 = (x**2 for x in range(3))   # 소괄호 → generator (제너레이터)
squares2 = [x**2 for x in range(3)]   # 대괄호 → list (리스트)

print(type(squares1))   # <class 'generator'>
print(type(squares2))   # <class 'list'>

print(squares1)   # <generator object <genexpr> at 0x...>  ← 값이 안 보임!
print(squares2)   # [0, 1, 4]  ← 값이 바로 보임

### 리스트 컴프리헨션([])
# **리스트 컴프리헨션([])**은 즉시 모든 값을 계산해서 리스트로 생성
# 그래서 print하면 바로 [0, 1, 4]가 보여요.

### 제너레이터 표현식(())
# **제너레이터 표현식(())**은 값을 미리 다 만들어두지 않음.
# "필요할 때마다 하나씩 계산해서 줄게"라는 **약속**만 갖고 있는 상태
# 그래서 print하면 실제 값이 아니라 제너레이터 객체 자체가 나와요.

# 값을 꺼내려면 (제너레이터는 for나 next()로 하나씩 꺼내야 함)
squares1 = (x**2 for x in range(3))

for val in squares1:
    print(val)   # 0, 1, 4  ← for문 돌리면 순서대로 하나씩 나옴
    
squares1 = (x**2 for x in range(3))
print(next(squares1))   # 0  ← 하나씩 직접 꺼낼 수도 있음
print(next(squares1))   # 1
print(next(squares1))   # 4
print(next(squares1))   # StopIteration 에러! (다 썼음)


# 왜 이런 차이가 있는 거고, 언제 뭘 써야 하나 (메모리 차이)
# 이게 진짜 핵심! 제너레이터는 메모리를 훨씬 아낌
# list: range(1000000)개 값을 전부 미리 계산해서 메모리에 저장 (무거움)
big_list = [x**2 for x in range(1000000)]

# generator: 값을 미리 만들지 않고, 필요할 때 하나씩만 계산 (가벼움)
big_gen = (x**2 for x in range(1000000))
````

반복문에서의 yield 사용
  - for문 내부에서는 자동으로 next()를 호출하고, StopIteration이 발생하면 반복 종료합니다.

````python
def count_up_t(n):
    i = 1
    while i <= n:
        yield i     # yield를 사용하면 generator로 변환하겠다는 의미임
        i += 1
        print(i)

data = count_up_t(3)
print(data)

next(data)
next(data)
next(data)
next(data)
````

````python
# 만약 1,000만 개의 데이터 중 연산 비용이 아주 큰 작업(예: time.sleep이 포함되거나 복잡한 수학 연산)을 수행하다가, 
# 원하는 조건(예: 3의 배수)을 가진 첫 번째 결과만 찾고 싶을 때 두 방식의 차이

import time


# 무겁고 오래 걸리는 연산을 시뮬레이션하는 함수
def heavy_processing(n):
    # 실제로는 아주 복잡한 연산이 일어난다고 가정합니다.
    return n * 2


# 1. 조기 평가 (Eager Evaluation) 방식
def eager_approach(data):
    # 조건과 상관없이 1,000만 개 데이터 전체에 대해 heavy_processing을 먼저 '전부' 실행함
    result = [heavy_processing(x) for x in data]
   
    # 그 후, 필요한 조건을 찾음
    for val in result:
        if val > 0 and val % 3 == 0:
            return val


# 2. 지연 평가 (Lazy Evaluation) 방식
def lazy_approach(data):
    # 이 시점에는 아무런 연산도 하지 않음 (계산식만 준비)
    result_generator = (heavy_processing(x) for x in data)
   
    # 필요한 순간(next)에만 heavy_processing을 하나씩 실행함
    for val in result_generator:
        if val > 0 and val % 3 == 0:
            return val


# 테스트 데이터 (0부터 9,999,999까지)
data_list = range(10000000)


# [비교 시작]
# 1) 조기 평가는 1,000만 번의 heavy_processing을 모두 수행하느라 엄청난 시간이 걸림
# eager_approach(data_list)
start_time = time.time()
print(f"지연 평가 결과: {eager_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")


# 2) 지연 평가는 단 4번(x=0,1,2,3)만 연산하고 x=3일 때 6(3의 배수)을 반환하며 즉시 종료!
start_time_2 = time.time()
print(f"지연 평가 결과: {lazy_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time_2:.6f}초")
````

## any()
* **"여러 개 중에 하나라도 참(True)이면 전체를 True로 판단"**해주는 내장 함수예요.

````python
# 기본사용법
print(any([False, False, True]))   # True  ← 하나라도 True가 있으니까
print(any([False, False, False]))  # False ← 전부 False니까
print(any([]))                      # False ← 빈 리스트는 무조건 False

# 실무에서 자주 쓰는 패턴: 조건식이랑 같이 씀
# 리스트를 그대로 넣기보다, 조건을 걸어서 판단하는 경우가 훨씬 많아요.
scores = [45, 60, 30, 55]

# 60점 이상인 사람이 한 명이라도 있는지 확인
print(any(score >= 60 for score in scores))   # True (60점이 있으니까)
````

````python
# 예제 1: 뉴스 기사 요약용 키워드 필터링기 (Comprehension 활용)
# - 목표: 주어진 뉴스 본문 리스트에서 특정 키워드를 포함한 문장만 필터링

news_articles = [
    "ChatGPT is transforming the AI world.",
    "Bitcoin price hits a new high.",
    "Python dominates machine learning.",
    "Weather forecast shows heavy rain.",
]

keywords = ['AI', 'Python', 'machine learning']

filtered = [article for article in news_articles if any(kw in article for kw in keywords)]
print(filtered)
````
