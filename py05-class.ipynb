````python
class Person:
    def __init__(self, name, age, job):
        self.name = name    # self.멤버변수 = 매개변수
        self.age = age
        self.job = job

    def eat(self): 
        print('eat')

    def walk(self):
        print('walk')

    def print_info(self):
        print(f"{self.name}: {self.age}")

hong = Person('홍길동',30,'학생')
hong.eat()
hong.walk()
hong.print_info()

lee = Person('이영희', 25, '개발자')
lee.eat( )
lee.walk()
print(lee.name, lee.age, lee.job)
````

````python
class FourCal():
    def __init__(self, a, b):
        print('init')
        self.a = a
        self.b = b

    def add(self):
        # 클래스의 멤버변수를 베소드를 호출해서 사용할 수 있다
        return self.a + self.b
    
    def sub(self):
        return self.a - self.b
    
    def mul(self):
        return self.a * self.b
    
    def div(self):
        return self.a // self.b
````

````python
fc = FourCal(4, 2) # 생성자는 객체의 초기값을 할당한다.

print(fc.add())
print(fc.sub())
print(fc.mul())
print(fc.div())
````

````python
def x(fun):                 # fun이라는 매개변수로 "함수 자체"를 받음
    def y(x):               # x 함수 안에 또 다른 함수 y를 정의
        return fun(x+1)     # y 안에서 바깥의 fun을 사용
    return y                # y라는 함수 자체를 리턴함 (실행 결과가 아니라 함수 그 자체!)
````

````python
def x(fun):
    def y(x):
        return fun(x+1)
    return y
````

````python
# @를 붙이면 decorator임
@x
def s(a):
    return a

# s()함수에 @x 데코레이터가 선언되어 있지 않을 때 호출하는 방법
print(x(s)(5))

# s()함수에 @x 데코레이터가 선언되어 있을 때 호출하는 방법
print(s(5))
````

### classmethod

- 클래스를 이용해 무언가를 해야할때 사용한다.

````python
class B:
  x = 3          # 클래스 변수
  @classmethod
  def aa(cls):  # classmethod는 관례상 cls를 사용한다.
    print('python')

  def bb(self):
    print('instance')
````

````python
# 클래스.클래스변수
print(B.x)

# 클래스.클래스메소드
print(B.aa())

# self가 받는 값이 필요함
# TypeError: B.bb() missing 1 required positional argument: 'self'
# print(B.bb())

ab = B()
print(B.bb(ab))
````

### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.


| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |

````python
class B:
  x = 1

  def xx(self, x): # 인스턴스 메소드
    self.x = x

  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x

  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a + b)

  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
````

#### + 기호 사용 시 public
+setter & getter  
+toString():str 
```
[출력결과]
파이썬의정석의 저자는 김파이썬이고 150페이지로 구성됨
책이름 :  파이썬 입문
책저자 :  박북스
페이지 :  200 페이지
파이썬 입문의 저자는 박북스이고 200페이지로 구성됨
```


````python
class Book:
    def __init__(self, title=None, author=None, page=None):
        self.__title = title    # __(언더바2개)의 뜻은 private처럼 사용하라는 뜻으로 붙임 (직접 사용할 수 없게 자체 차단)
        self.__author = author
        self.__page = page

    def getTitle(self):
        return self.__title     # 외부에서 객체생성 후 사용 가능
    
    def setTitle(self, title):
        self.__title = title    # 객체 생성 후 title의 값을 바꿀 때

    def getAuthor(self):
        return self.__author

    def setAuthor(self, author):
        self.__author = author

    def getPage(self):
        return self.__page

    def setPage(self, page):
        self.__page = page

    def toString(self):
        return f"{self.__title}의 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
````

````python
class BookRun:
    def process(self):
        bk1 = Book()
        bk1.setTitle('파이썬의 정석')
        bk1.setAuthor('김파이썬')
        bk1.setPage(150)
        print(bk1.toString())

        # bk1.__title
        # __ 붙인건 _Book__title 이런식으로 생성이 됨
        print(bk1._Book__title)
        print(dir(bk1))

        bk2 = Book('파이썬 입문', '박북스', 200)
        print(f'책이름 : {bk2.getTitle()}')
        print(f'책저자 : {bk2.getAuthor()}')
        print(f'페이지 : {bk2.getPage()}')
        print(bk2.toString())


if __name__ == '__main__':
    bn = BookRun()
    bn.process()
````

클래스 메서드로 인스턴스 수 세기

````python
class Dog:
    count = 0

    def __init__(self, name):
        self.name = name
        Dog.count += 1

    @classmethod
    def get_count(cls):
        return cls.count
````

````python
class Dog:
    count = 0

    def __init__(self, name):
        self.name = name
        Dog.count += 1

    @classmethod
    def get_count(cls):
        return cls.count
    
    @staticmethod
    def is_valid_name(name):
        # name의 클래스가 문자열이고 길이는 0보다 큰지 확인
        return isinstance(name, str) and len(name) > 0
    
    
````

## 상속(inheritance)
 - class에서 이름 뒤에 괄호는 상속이다.
 - python의 모든 객체는 object를 상속받는다.
 - python3부터는 (object)는 생략 가능하다. 즉 지금까지 class를 정의할때 object를 내부적으로 상속하고 있었다.

```
class 도형:
 pass

class 삼각형(도형):  # 삼각형은 도형이다.
  pass

class 사각형(도형):  # 사각형은 도형이다.
  pass
```

````python
# 부모클래스
class Parent:
    x = 'parent x'  #클래스 변수
    y = 'parent y'  #클래스 변수


    @classmethod
    def output(cls):
        return cls.x


    def test(self):
        return 10


# 자식클래스
class Child(Parent):
    x = 'child x' # 클래스 변수


    def display(self):
        print('display')
        print(Parent.output())
        print(Child.output())
        print(Parent.test(self))
        print(Child.test(self))
````

````python
print(Parent.x)
print(Parent.y)

print(Child.x)
print(Child.y)

````

## overriding(오버라이딩)
부모가 가진 기능을 변경하는 것을 overriding이라고 한다.  
double underbar 같은 것들은 overriding해서 객체의 기능을 바꿀 수 없다.

````python
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        # 자식 클래스들이 반드시 재정의(오버라이딩)하도록 유도하는 부모의 기본 메서드
        return "알 수 없는 동물의 소리입니다."
    
# 자식 클래스 1 (Dog)
class Dog(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "멍멍!"


# 자식 클래스 2 (Cat)
class Cat(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "야옹~"
````

````python
animal = Animal('동물')
print(animal.make_sound())

dog = Dog('강쥐')
print(dog.make_sound())

cat = Cat('냥이')
print(cat.make_sound())
````

````python
class Base:
    def __init__(self):     # 8
        self.x = 'Base x'   # 9
        self.y = 'Base y'   # 10
        print('Base')       # 11


class A(Base):
    a = 1
    def __init__(self):     # 4
        super().__init__()  # 5
        self.x = 'A x'      # 14
        print('A')          # 15


class B(Base):
    b = 2
    def __init__(self):     # 6
        super().__init__()  # 7
        self.x = 'B x'      # 12
        print('B')          # 13


class C(A, B):
    def __init__(self):     # 1
        print('C init')     # 2
        #super(C, self).__init__()  # python 1, 2 버전에서 사용
        super().__init__()  # 3    # python 3에서 사용
        print('C')          # 16

# 상속관계의 순서 확인
print(C.__mro__)
````

````python
from abc import ABC, abstractmethod

# Shape는 설계도일 뿐이야. area()를 실제로 구현하지 않은 채로는 절대 객체로 만들 수 없어
# Shape는 "사용하는" 클래스가 아니라 "상속받는" 클래스임
# Shape를 직접 호출하는 게 아니라, 
# Shape를 상속받은 자식 클래스를 만들고, 그 자식이 area()를 반드시 구현하게 강제하는 용도로 사용
class Shape(ABC):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    @abstractmethod
    def area(self):
      pass    

class Rect(Shape):
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height
````

````python
s = Shape(3, 2)
````

````python
r = Rect(2, 5)
print(f"사각형 넓이: {r.area()}")
````
